```python
#VADER
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def sentiment_vader(sentence):

    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()

    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
    negative = sentiment_dict['neg']
    neutral = sentiment_dict['neu']
    positive = sentiment_dict['pos']
    compound = sentiment_dict['compound']

    # print("Overall sentiment dictionary is : ", sentiment_dict)
    # print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
    # print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
    # print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
    #
    # print("Sentence Overall Rated As", end = " ")
    #
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        overall_sentiment = "Positive"
        # print("Positive")

    elif sentiment_dict['compound'] <= - 0.05 :
        overall_sentiment = "Negative"
        # print("Negative")

    else :
        overall_sentiment = "Neutral"
        # print("Neutral")
    #
    return negative, neutral, positive, compound, overall_sentiment
```

```python
data['sentiment_negative', 'sentiment_neutral', 'sentiment_positive','compound','overall_sentiment'] = data.headline_text.map(sentiment_vader)
```

```python
data.to_excel('data_sentiment_vader.xlsx', encoding = 'utf8')
```

```python
#TextBlob
from textblob import TextBlob
```

```python
def sentiment_texblob(row):
    classifier = TextBlob(row)
    # polarity = str(classifier.sentiment.polarity)
    # subjectivity = str(classifier.sentiment.subjectivity)

    polarity = classifier.sentiment.polarity
    subjectivity = classifier.sentiment.subjectivity
    print(polarity)
    print(subjectivity)

    return polarity,subjectivity
```

```python
#Happy Transformer
from happytransformer import HappyTextClassification
```

```python

def sentiment_ht(row):
    happy_tc = HappyTextClassification("BERT", "ProsusAI/finbert", num_labels=3)

    result = happy_tc.classify_text(row)
    label = result.label
    score = result.score
    print(row)

    return label, score
```
